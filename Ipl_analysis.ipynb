```python
def convert_to_datetime(date_str):
    try:
        # Try parsing with the first format
        return pd.to_datetime(date_str, format='%Y-%m-%d')
    except ValueError:
        try:
            # Try parsing with the second format
            return pd.to_datetime(date_str, format='%m/%d/%y')
        except ValueError:
            # If both formats fail, return NaT
            return pd.NaT

# Apply the custom function to each element in the date_data
merged_df['date'] = pd.Series(merged_df['date']).apply(convert_to_datetime)
```
